```python
import pandas as pd
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("tarifas metro.xlsx")
data
```

```python
colnames = data.columns.values.tolist()
estaciones = colnames[2:3]
tarifa = colnames[1:2]
Y = data[tarifa]
X = data[estaciones]
```

```python
tarifaa = np.array([Y], dtype=float)
estacioness = np.array([X], dtype=float)
```

```python
oculta1 = tf.keras.layers.Dense(units=50, input_shape=[1])
oculta2 = tf.keras.layers.Dense(units=50)
salida = tf.keras.layers.Dense(units=1)
modelo = tf.keras.Sequential([oculta1, oculta2, salida])
```

```python
modelo.compile(optimizer=tf.keras.optimizers.Adam(),loss="mean_squared_error")
```

```python
historial = modelo.fit(X,Y, epochs=1000)
```

```python
resultado = modelo.predict(X)
```

```python
resultado
```

```python
data["pred tarifa"] = resultado
data
```

```python
real_stock_price = data.iloc[:, 1:2].values
# Visualizando Resultados
plt.plot(real_stock_price, color = 'red', label = 'tarifa')
plt.plot(resultado, color = 'blue', label = 'pred tarifa')
plt.title('Tarifa metro Santiago')
plt.xlabel('estaciones')
plt.ylabel('tarfia')
plt.legend()
plt.show()
```
