# Importar librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

# Dataset

```python
data = pd.read_csv("Desktop/Precio café.csv")
data # Este dataset trata sobre el precio del café desde 1969 hasta 2022 (dólares por libra de café)
```

# Entrenamiento predicción

```python
colnames = data.columns.values.tolist()
predictors = colnames[:13]
target = colnames[2]
X = data[predictors]
Y = data[target]
```

# Árbol de regresión

```python
arbolregr = DecisionTreeRegressor(min_samples_split=30, min_samples_leaf=10, max_depth=5, random_state=0)
```

```python
arbolregr.fit(X,Y)
```

```python
Predicción = arbolregr.predict(data[predictors])
```

# Predicción v/s Precio final

```python
data["Predicción"] = Predicción
```

```python
data[["Predicción", "Precio final"]]
```

# Estadísticas

```python
data.describe()
# Aquí se puede apreciar el promedio del precio final, y la predicción. Se puede ver que la diferencia de precio es de 
# 0,01789 (equivalente a 17,03 pesos chilenos), haciendo la predicción muy cercana a la realidad.
```
