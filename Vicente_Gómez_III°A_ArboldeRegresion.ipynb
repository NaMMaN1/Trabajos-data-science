# Importar librer√≠as

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

# Dataset

```python
data = pd.read_csv("Desktop/Precio caf√©.csv")
data # Este dataset trata sobre el precio del caf√© desde 1969 hasta 2022 (d√≥lares por libra de caf√©)
```

# Entrenamiento predicci√≥n

```python
colnames = data.columns.values.tolist()
predictors = colnames[:13]
target = colnames[2]
X = data[predictors]
Y = data[target]
```

# √Årbol de regresi√≥n

```python
arbolregr = DecisionTreeRegressor(min_samples_split=30, min_samples_leaf=10, max_depth=5, random_state=0)
```

```python
arbolregr.fit(X,Y)
```

```python
Predicci√≥n = arbolregr.predict(data[predictors])
```

# Predicci√≥n v/s Precio final

```python
data["Predicci√≥n"] = Predicci√≥n
```

```python
data[["Predicci√≥n", "Precio final"]]
```

```python
data.describe()
# Aqu√≠ se puede apreciar el promedio del precio final, y la predicci√≥n. Se puede ver que la diferencia de precio es de 
# 0,01789 (equivalente a 17,03 pesos chilenos), haciendo la predicci√≥n muy cercana a la realidad.
```
